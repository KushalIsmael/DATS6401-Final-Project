```python
import numpy as np
pivot = pd.pivot_table(party, values='Count', index=['State'], columns = ['Party'], aggfunc=np.sum, fill_value=0)
```

```python
pivot
```

```python
pivot.info()
```

```python
pivot = pivot.drop(columns=['Independent','Libertarian'])
```

```python
pivot['R-D']= pivot['Republican']-pivot['Democratic']
```

```python
pivot
```

```python
pivot.loc[pivot['R-D'] > 0, 'Majority Party'] = 'Republican'
pivot.loc[pivot['R-D'] < 0, 'Majority Party'] = 'Democrat'
pivot.loc[pivot['R-D'] == 0, 'Majority Party'] = 'No Majority'
```

```python
pivot
```

```python
pivot.to_csv('Political Party.csv')
```
